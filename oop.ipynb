## Trụ cột OOP:
1. Encapsulation: Đóng gói attributes & methods, ẩn chi tiết nội bộ
    - Scope of accessing:
        - `name`: (no `_`): public: can access everywhere
        - `_name`: (1 `_`): protected (__convention only__): can access in class + subclass
        - `__name`: (2 `_`): private: only accessible in class
2. Inheritance: Class con kế thừa attr & methods từ class cha
3. Polymorphism: Cùng tên method nhưng có hành vi khác nhau ở từng class
4. Abstraction: Ẩn độ phức tạp, chỉ hiện thị những gì cần thiết

## Special Methods (for overiding):
- `__init__`
- `__str__`
- `__eq__`
- `__len__`

```python
class SinhVien:
    # Class attribute (dùng chung cho mọi objects created with this class)
    college = "NEU"
    def __init__(self, fullname, msv, score):
        # Instance attributes (unique for each object)
        self.fullname = fullname
        self.msv = msv
        self.score = score
    
    # Instace method
    def rank(self):
        if self.score >= 8.5:
            return "XS"
        elif self.score >=7.0:
            return "Good"
        else:
            return "Normal"
        
    def __str__(self):
        return f"Name: {self.fullname} - {self.msv} is a {self.rank()} student"
```

## Encapsulation

```python
class Test():
    def __init__(self):
        self.a = "a" # Public
        self._b = "b" # Protected (Convention)
        self.__c = "c" # Private
        
t = Test()
print(t.a)
print(t._b)
print(t.__c)
```

## Inheritance

```python
class NhanVien:
    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary
    
    @property # Convert method to attribute
    def showSalary(self):
        return self.__salary

    def calcSalary(self):
        return self.__salary

    def show(self):
        pass
        # print(f"NV: {self.name} | Salary: {self.}")
        
        
# Children class
class Manager(NhanVien):
    def __init__(self, name, salary, specialty):
        super().__init__(name, salary) # Call the parent's __init__
        self.specialty = specialty
        self.bonus = 0.4
        
    def calcSalary(self): # Overide the parent's method
        return self.__salary + self.bonus
    
class Worker(NhanVien):
    def __init__(self, name, salary, falcuty):
        super().__init__(name, salary)
        self.falcuty = falcuty
        self.bonus = 0.2
        
    def calcSalary(self):
        return super().calcSalary() + self.__salary * self.bonus
    
    
manager1 = Manager("A", 2000, "IT")
nv1 = Worker("A1", 200, "IT")
```

## Abstraction

```python
# abc: Abstract Base Class
from abc import ABC, abstractmethod

class Shape(ABC): # Abstract Class
    @abstractmethod
    def area(): # Abstract method pre-defined to overide
        pass
    
    @abstractmethod
    def circum():
        pass
    
    
class Circle(Shape):
    def __init__(self, radius):
        self.r = radius
    
    def area(self):
        return 3.14 * self.r ** 2
    
    def circum(self):
        return 2 * 3.14 * self.r
    
class Rect(Shape):
    def __init__(self, h, w):
        self.h = h
        self.w = w
    
    def area(self):
        return self.w * self.h
    
    def circum(self):
        return 2*(self.w+self.h)
    

circ1 = Circle(10)
print(circ1.area())

rect1 = Rect(10, 20)
print(rect1.area())

# Abstract methods need to be overiden
shape1 = Shape() # Can't instantiate abstract class Shape with abstract methods area, circum
```
